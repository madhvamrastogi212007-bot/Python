```python
#SQUARE HOLLOW PATTERN
n=int(input("ENTER A NUMBER FROM A USER"))
for i in range(1,n+1):#i REPRESENTS ROWS
    for j in range(1,n+1):#j REPRESENTS COLUMNS
        if(i==1 or i==n or j==1 or j==5):
            print("*",end="  ")
        else:
            print(" ",end="  ")
    print()
```

```python
#NUMBER TRIANGULAR
n=int(input("ENTER A NUMBER OF ROWS"))
for i in range(n):
    for j in range(n-i-1):#TO PRINT THE SPACES,6-0-1=5-SPACES | 6-1-1=4-SPACES
        print(" ",end="")
    for j in range(i+1):#TO PRINT THE STARS
        print(i,end=" ")
    print()
```

```python
#NUMBER INCREASING PYRAMIDS
n=int(input("ENTER A NUMBER OF ROWS"))
for i in range(1,n):#rows-IT REPRESENTS OUTER LOOP
    for j in range(1,i+1):#column-IT RPRESENTS INNER LOOP
        print(j,end=" ")
    print()
```

```python
#NUMBER INCREASING PYRAMID
for i in range(4,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#NUMBER CHANGING PATTERN
n=4
num=1
for i in range(1,n+1):
    for j in range(1,i+1):
        print(num,end=" ")#PRINT THE CURRENT VALUE OF NUM i.e-1
        num+=1# INCREASE NUMBER BY 1 /// THEN NUM BECOME NUM=1+1=2,2+1=3.....
    print()
```

```python
#NUMBER TRIANGULAR
n=int(input("ENTER A NUMBER OF ROWS"))
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range(2*i+1):
        print(i,end=" ")
    print()
```
