```python
# Hollow Triangle
rows = int(input("Enter number of rows: "))

for i in range(1, rows+1):
    # Print spaces
    for j in range(rows-i):
        print(" ", end="")
    
    # Print stars and spaces inside
    for j in range(1, 2*i):
        if j == 1 or j == 2*i-1 or i == rows:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
