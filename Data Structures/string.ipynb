- what does symmetric means?
   - when someting is divided in two equal parts, both parts looks exactly the same
     
      --A STRING IS SYMMETRIC IF:
         - length of string is even
         - when you split it  from the middle
         - when first_half==second_half

```python
# Check Whether the String is Symmetrical or Palindrome - Python
# [::-1] means read the string from back to front

# palindrome
a="madam"
if a==a[::-1]:
    print(a," is a palindrome ")
else:
    print("not palindrome ")


length = len(a)

if length % 2 == 0:
    mid = length // 2
    first_half = a[:mid]
    second_half = a[mid:]

    if first_half == second_half:
        print(a,"is a Symmetric")
    else:
        print(a,"is Not Symmetric")
else:
    print(a,"is not symmetric") # bcoz if we will not use this then it will not show
    # a,"is not symmetric" print this if len is odd
    # print this (a,"is a Symmetric") if the len is even
```

-enumerate() is a built-in Python function that adds a counter (index) to an iterable and returns it as an enumerate object.
 - This allows you to loop over the iterable and access both the index and the value at the same time.
  -- key point:
      - works with strings,lists,tuples and other iterable
      - default value start by 0,but you can change it with start parameter
      - useful when position of elements matter
(iterable) mean -Iterable is a Python object that can be looped over, giving one element at a time.

```python
a="madam"
for i,char in enumerate(a):
    print(i,char)
```

```python
#python program to capitalize the first and last character of each word in a string
# Strings are immutable in Python, so to modify characters we create new strings. 
# Lists are mutable, so we temporarily store modified data in a list and finally join it back into a string.
n="bill gates"
s1=n.split()
# Step 2: Create an empty list to store modified words
# We use a list because strings are immutable
result=[]
# If the word has only one character
# then simply convert it to uppercase
for w in s1:
    if len(w)==1:
        result.appernd(w)
    else:
# Capitalize the first character
# Keep the middle characters same
# Capitalize the last character
        new_word=w[0].upper() + w[1:-1] + w[-1].upper()
        result.append(new_word)

final_string=' '.join(result) #Join the list of words back into a string
print(final_string)
```
