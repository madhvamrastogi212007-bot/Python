-SLICING

```python
apps=["instagram","trading view","zerodha","angleone","grow","whatsapp","facebook","money control","snapchat"]
print(apps[2:8])
print(apps[1:10:2])
print(apps[:]) # it will print whole list
print(apps[2:])# it will print from starting position(2) till the end
print(apps[:6])# it will print till 5
print(apps[::2])# by taking gap of 2  print
print(apps[-6:-2]) # by using - it will always take from left to right & will print till -1
```

LIST METHODS

- append()-to add item at the end of the list
- insert()- to add item at any specific positions in the list
- pop()- to delete item using its index position
- remove()- to delete item  using  item name 
- copy()- to make a duplicate of a list
- sort()- to sort a list in a particular order
-  extend()- to combine two list in a single list 
- clear()- to delete all the item from the lists but not the list 

```python
movies=['Your Name','AOT','Dhurandhar','Your Fault','Overdose','Bahubali']
movies.append('Matrix')
print(movies)
movies.insert(2,'INTERESTELLER')
movies
```

```python
movies=['Your Name','AOT','Dhurandhar','Your Fault','Overdose','Bahubali']
movies.pop()# removes last element   and first element and also modifies
print(movies)
movies.pop(0)
```

```python
movies=['Your Name','AOT','Dhurandhar','Your Fault','Overdose','Bahubali']
movies.remove(movies[1]) # to remove elements using index value
print(movies)
```
