```python
#CLONING OR COPYING OF LIST
b=[2,7,0,3,75,58,9,1,3,-3,-19,[1, 2], [3, 4]]
# Creating a shallow copy of list b and hence
#shallow copy does NOT copy nested objects.
#It only copies the references of inner objects.
# Only the outer list is copied, nested lists are shared
c= b.copy()
# Printing the copied list (same content as b initially)
print(c)
# Modifying the first nested list in the copied list
# Since this is a shallow copy, the inner list is the SAME object in both b and c
c[11].append(20)
print("Outer changes")
print(b)
print("original: ",c)
```

```python
#Program to Print Duplicates from a List of Integers in Python

nums = [1, 2, 3, 2, 4, 1, 5, 2]

# Empty list to store duplicate elements
duplicates = []

# Loop through each element in the list
for i in nums:
    
    # Check if the element appears more than once in the list
    # AND it is not already added to the duplicates list
    if nums.count(i) > 1 and i not in duplicates:
        
        # Add the duplicate element to the duplicates list
        duplicates.append(i)


print("Duplicate elements are:", duplicates)

```

```python
#Get Unique Values from a List in Python
nums = [1, 2, 3, 2, 4, 1, 5, 2]

# Empty list to store duplicate elements
unique=[]

# Loop through each element in the list
for num in nums:
    
    # Check if the element is not already present in unique list
    if num not in unique:
        
        # Add the element only once otherwise skip if it is in included in unique
        unique.append(num)

print("unique elements are:", unique)

```

```python
#  Strong Number
#A number whose sum of factorial of digits equals the number. example-145 → 1! + 4! + 5! = 1 + 24 + 120 = 145
num=145
temp=num
s=0
while(temp>0):

    rem=temp%10
    fact=1    # # reset factorial for EACH digit & factorial calculation must start fresh for each digit

    for i in range(1,rem+1):
        fact=fact*i
    s=s+fact
    temp=temp//10    

if(s==num):
   print(s," strong number")
else:
   print(s," not strong number")

        
```

```python
# PALINDROME NUMBERS
n=int(input("ENTER A NUMBER"))
orig=n
rev=0

while n>0:
    digits=n%10
    rev=rev*10 + digits
    n=n//10

if orig==rev:
    print(f"{orig} is a palindrome number.")
else:
    print(f"{orig} is not a palindrome number")
```
