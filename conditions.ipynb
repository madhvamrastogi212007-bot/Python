```python
side1=10
side2=8
side3=12
if (side1==side2==side3):
    print("EQUILATERAl triangle")
elif side1==side2 or side2==side3 or side1==side3:
    print("ISOCELESS TRIANGLE")
else:
    print("SCALENE TRIANGLE")     
```

```python
#IF WILL CHECK THE CONDITIONS AT ALL LEVEL EVEN IF THE ONE CONDITION BECOMES TRUE AND ELSE WILL STOP IF  even  after the one CONDITION BECOMES TRUE
side1=10
side2=10
side3=10
if (side1==side2==side3):
    print("EQUILATERAl TRIANGLE")
if side1==side2 or side2==side3 or side1==side3:
    print("ISOCELES TRIANGLE")
if side1!=side2 or side2!=side3 or side1!=side3:
    print("SCALENE TRIANGLE")    
```
