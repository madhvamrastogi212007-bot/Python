- FOR LOOPS

```python
fruits=["apple","banana","grapes","orange"]
for x in fruits:#X WILL CHECK AGAIN AND AGAIN UNTIL THE  ELEMENTS IN LISTS GOT DISPLAYED
    print(x,end=" ")
```

```python
name="madhav rastogi"
for x in name:
    print(x,end="")
```

```python
for i in range(1,3):
    print(i)
```

```python
#PRINT YOUR NAME 6 TIMES
name=["madhav","madhav","madhav","madhav","madhav","madhav"]
for i in name:
    print(i,end=" ")
```

```python
#PRINT ALL EVEN NUMBERS TILL 20 WITHOUT USING CONDITION STATEMENT 
for i in range(0,21,2):
    print(i)
```

```python
#PRINT ALL ODD NUMBERS TILL 20 USING CONDITION STATEMENT
for i in range(1,21):
    if i%2!=0:
        print(i)
```

- WHILE LOOP

```python
x=0
while x<=5:
    print(x,end=" ")
    x+=1
```

```python
x=1
while x<3:
    print(x,end=" ")
    x+=1
```

```python
#PRINT YOUR NAME 6 TIMES
a="madhav"
b=0
while b<=5:
    print(a)
    b=b+1
```

```python
#PRINT ALL EVEN NUMBERS TILL 20 WITHOUT USING CONDITION STATEMENT 
x=0
while x<21:
    x%2==0
    print(x,end=" ")
    x+=2
```

```python
x=1
while x<21:
    x%2!=0
    print(x,end=" ")
    x+=1
```
